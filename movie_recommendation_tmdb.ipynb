```python
import pandas as pd
import numpy as np
import ast #to convert string to list
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies=movies[['title','movie_id','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
def convert_genre(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert_genre)
```

```python
movies['keywords']=movies['keywords'].apply(convert_genre)
```

```python
def convert_cast(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.iloc[0]['crew']
```

```python
def convert_crew(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

As dimension goes up eclidian distance failes the most use cosine distance instead i.e. calculate the angle between the two vectors

in pca mean centering is very important

distance in inv prop to similartity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index=int(new_df[new_df['title']==movie].index[0])
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
